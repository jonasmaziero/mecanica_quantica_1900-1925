# Equação de Schrödinger
Essa equação é uma das equações mais importantes de toda a história da humanidade, e tem a forma
\begin{equation}
\color{blue}{i\hbar\partial_{t}\Psi(\vec{r},t)=-\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}\Psi(\vec{r},t)+U(\vec{r},t)\Psi(\vec{r},t)},
\end{equation}
com $\partial_{t}\equiv\frac{\partial}{\partial_{t}}$, em coordenadas cartesianas $\nabla^{2}_{\vec{r}}=\partial_{xx}+\partial_{yy}+\partial_{zz}$ e também podemos usar a notação $\Psi(\vec{r},t)=\Psi(x,y,z,t)$.

OBS. Note $i=\sqrt{-1}$, que não aparece na eq. de ondas clássica, faz parte da equação de Schrödinger (EqS). Por isso, em geral $\Psi(\vec{r},t):\mathbb{R}^{4}\rightarrow\mathbb{C}$.

OBS. Note que a EqS envolve a _derivada primeira_ no tempo e derivadas segundas no espaço, o que contrasta com a eq. de ondas clássica,
\begin{equation}
v^{2}\nabla^{2}_{\vec{r}}f=\partial_{tt}f,
\end{equation}
que envolve derivadas segundas no tempo e no espaço.

OBS. A EqS, assim como a segunda lei de Newton da Mecânica Clássica e as equações de Einstein da Relatividade Geral, _não pode ser deduzida_. No entando, podemos dar uma certa plausibilidade para ela, como faremos na sequência.

OBS. A EqS é a base da Mecânica Quântica não relativística, e a princípio só se aplica nesse regime de movimento, $v\ll c$.

#### Tornando a EqS plausível
Para um quanton no limite não relativístico de baixas velocidades, se usamos as relações de de Broglie-Einstein,
\begin{equation}
E=h\nu=\hbar\omega \text{ e } p=\frac{h}{\lambda}=\hbar k,
\end{equation}
a equação para a energia do quanton fica escrita como:
\begin{equation}
E=\frac{P^{2}}{2m}+U \text{ ou } \hbar\omega=\frac{\hbar^{2}k^{2}}{2m}+U.
\end{equation}
Agora, se considerarmos uma onda harmônica escrita na forma complexa
\begin{equation}
\Psi=\Psi(x,t)=Ae^{i(kx-\omega t)},
\end{equation}
então $\partial_{t}\Psi(x,t)=-i\omega\Psi(x,t)$, $\partial_{xx}\Psi(x,t)=i^{2}k^{2}\Psi(x,t)=-k^{2}\Psi(x,t)$ e $\partial_{yy}\Psi(x,t)+\partial_{zz}\Psi(x,t)=0$. Substituindo na ES teremos
\begin{align}
& i\hbar\partial_{t}\Psi(x,t)=i\hbar(-i\omega\Psi(x,t))=\hbar\omega\Psi(x,t) \\
& = -\frac{\hbar^{2}}{2m}\partial_{xx}\Psi(x,t)+0+0+U\Psi(x,t) = \frac{\hbar^{2}}{2m}k^{2}\Psi(x,t)+U\Psi(x,t),
\end{align}
que nos dá $\hbar\omega=\hbar^{2}k^{2}/2m + U$, como esperado.

## EqS independente do tempo
Os sistemas físicos que estudaremos nessa disciplina serão, em geral, aproximados de forma que a energia potencial deles não depende do tempo, i.e.,
\begin{equation}
U=U(\vec{r}).
\end{equation}
Nessas condições podemos separar a EqS em duas eqs. diferenciais, uma envolvendo o tempo e a outra o espaço fazendo _separação de variáveis_:
\begin{equation}
\Psi(\vec{r},t)=\psi(\vec{r})\phi(t).
\end{equation}
Teremos assim que
\begin{align}
& i\hbar\partial_{t}(\psi(\vec{r})\phi(t)) = \psi(\vec{r})i\hbar\partial_{t}\phi(t) \\
& = -\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}(\psi(\vec{r})\phi(t))+U(\vec{r})(\psi(\vec{r})\phi(t)) = -\phi(t)\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}\psi(\vec{r})+U(\vec{r})(\psi(\vec{r})\phi(t)).
\end{align}
Dividimos essa equação por $\psi(\vec{r})\phi(t)$ para obter
\begin{align}
\frac{i\hbar\partial_{t}\phi(t)}{\phi(t)} = -\frac{1}{\psi(\vec{r})}\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}\psi(\vec{r})+U(\vec{r}).
\end{align}

Como $\vec{r}$ e $t$ são variáveis independentes, mudanças em um lado da eq. acima não podem afetar o outro lado. Portanto os dois lados dessa eq. tem que ser iguais a uma mesma constante:
\begin{equation}
i\hbar\partial_{t}\phi(t)=C\phi(t) \therefore \partial_{t}\phi(t)=-(iC/\hbar)\phi(t) \therefore \phi(t)=e^{-iCt/\hbar}.
\end{equation}
Podemos inferir que a constante é igual a energia usando o mesmo exemplo de função de onda usado anteriormente:
\begin{equation}
\Psi(x,t) = Ae^{i(kx-\omega t)} = Ae^{ikx}e^{-i\omega t} =: \psi(x)\phi(t).
\end{equation}
Assim
\begin{equation}
i\hbar\partial_{t}\phi(t) = i\hbar(-i\omega)\phi(t) = \hbar\omega\phi(t) = E\phi(t)=C\phi(t).
\end{equation}
Por conseguinte
\begin{equation}
C=E \Rightarrow \phi(t)=e^{-iEt/\hbar}
\end{equation}
e
\begin{equation}
\color{blue}{-\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}\psi(\vec{r})+U(\vec{r})\psi(\vec{r}) = E\psi(\vec{r})}.
\end{equation}

OBS. Nas aplicações da Mecânica Quântica estaremos quase sempre preocupados em resolver essa equação, chamada de eq. de Schrödinger independente do tempo (EqSIT), para algumas dependências específicas da energia potencial com a posição.

OBS. Poderíamos inferir que $C=E$ notando que a EqSIT é uma equação de autovalores e autovetores para o operador Hamiltoniano do sistema:
\begin{equation}
\left(\frac{P^{2}}{2m}+U(\vec{r})\right)\psi(\vec{r}) = H\psi(\vec{r}) = E\psi(\vec{r}),
\end{equation}
onde usamos o operador momento linear (veremos o porquê disso na disciplina de Mecânica Quântica II [https://github.com/jonasmaziero/mecanica_quantica/blob/master/09_position.ipynb]):
\begin{equation}
P \equiv \frac{\hbar}{i}\vec{\nabla}_{\vec{r}}\therefore P^{2} = -\hbar^{2}\nabla^{2}_{\vec{r}}.
\end{equation}

# Interpretação da função de onda
A função de onda de um quanton, $\Psi(\vec{r},t):\mathbb{R}^{4}\rightarrow\mathbb{C}$, é uma função complexa. Por isso essa função não possui uma interpretação direta. Sua interpretação foi dada por Born [https://www.nobelprize.org/prizes/physics/1954/born/lecture/] e é a seguinte:
\begin{equation}
|\Psi(\vec{r},t)|^{2}d^{3}r = \text{probabilidade do quanton estar no volume } d^{3}r \text{ em torno da posição } \vec{r} \text{, no instante de tempo } t.
\end{equation}
A função de onda $\Psi(\vec{r},t)$ é uma _amplitude de probabilidade_, e sua ligação com quantidades físicas observáveis é dada pela interpretação de Born acima.

Dada essa interpretação, a função de onda precisa satisfazer a _condição de normalização_, ou seja, se, no instante $t$, somamos as probabilidades das posições do quanton por todo o espaço, temos que obter unidade (o quanton tem que estar em algum lugar):
\begin{equation}
\iiint|\Psi(\vec{r},t)|^{2}d^{3}r=1.
\end{equation}

OBS. Para $U=U(\vec{r})$ usaremos
\begin{equation}
|\Psi(\vec{r},t)|^{2} = |\psi(\vec{r})\phi(t)|^{2} = |\psi(\vec{r})|^{2}|e^{-iEt/\hbar}|^{2}=|\psi(\vec{r})|^{2}.
\end{equation}

# Valores esperados
OBS. Veremos de como as equações abaixo são obtidas em Mecânica Quântica II.

Os valores médios de posição e do seu quadrado são calculados usando
\begin{align}
\langle \vec{R}(t)\rangle & = \iiint \vec{r}|\Psi(\vec{r},t)|^{2}d^{3}r \\
& = \hat{i}\iiint x|\Psi(x,y,z,t)|^{2}dxdydz + \hat{j}\iiint y|\Psi(x,y,z,t)|^{2}dxdydz + \hat{k}\iiint z|\Psi(x,y,z,t)|^{2}dxdydz \\ 
& = \hat{i}\langle X(t)\rangle + \hat{j}\langle Y(t)\rangle + \hat{k}\langle Z(t)\rangle, \\
\langle R^{2}(t)\rangle & = \iiint r^{2} |\Psi(\vec{r},t)|^{2}d^{3}r \\
& = \iiint x^{2} |\Psi(x,y,z,t)|^{2}dxdydz + \iiint y^{2} |\Psi(x,y,z,t)|^{2}dxdydz + \iiint z^{2} |\Psi(x,y,z,t)|^{2}dxdydz \\ 
& = \langle X^{2}(t)\rangle + \langle Y^{2}(t)\rangle + \langle Z^{2}(t)\rangle, \\
\sigma_{X}(t) & = \sqrt{\langle R^{2}(t)\rangle-\langle \vec{R}(t)\rangle^{2}}.
\end{align}
OBS. Todas as integrais são feitas por todo o espaço. <br>
OBS. Se mudamos de sistema de coordenadas, usamos a mesma ideia.

Para cálculos de valores médios envolvendo o momento linear usamos
\begin{align}
\langle \vec{P}(t)\rangle & = \iiint \Psi(\vec{r},t)^{*}\vec{p}\Psi(\vec{r},t)d^{3}r = \iiint \Psi(\vec{r},t)^{*}\left(\frac{\hbar}{i}\vec{\nabla}_{\vec{r}}\right)\Psi(\vec{r},t)d^{3}r \\
& = \hat{i}\iiint\Psi(x,y,z,t)^{*}\left(\frac{\hbar}{i}\partial_{x}\right)\Psi(x,y,z,t)dxdydz + \hat{j}\iiint\Psi(x,y,z,t)^{*}\left(\frac{\hbar}{i}\partial_{y}\right)\Psi(x,y,z,t)dxdydz \\ 
& + \hat{k}\iiint\Psi(x,y,z,t)^{*}\left(\frac{\hbar}{i}\partial_{z}\right)\Psi(x,y,z,t)dxdydz \\ 
& = \hat{i}\langle P_{x}(t)\rangle + \hat{j}\langle P_{y}(t)\rangle + \hat{k}\langle P_{z}(t)\rangle, \\
\langle P^{2}(t)\rangle & = \iiint \Psi(\vec{r},t)^{*}p^{2}\Psi(\vec{r},t)d^{3}r = \iiint \Psi(\vec{r},t)^{*}\left(-\hbar^{2}\nabla^{2}_{\vec{r}}\right)\Psi(\vec{r},t)d^{3}r \\ \\
& = \iiint\Psi(x,y,z,t)^{*}\left(-\hbar^{2}\partial_{xx}\right)\Psi(x,y,z,t)dxdydz + \iiint\Psi(x,y,z,t)^{*}\left(-\hbar^{2}\partial_{yy}\right)\Psi(x,y,z,t)dxdydz \\ 
& + \iiint\Psi(x,y,z,t)^{*}\left(-\hbar^{2}\partial_{zz}\right)\Psi(x,y,z,t)dxdydz \\  
& = \langle P_{x}^{2}(t)\rangle + \langle P_{y}^{2}(t)\rangle + \langle P_{z}^{2}(t)\rangle, \\
\sigma_{P}(t) & = \sqrt{\langle P^{2}(t)\rangle-\langle \vec{P}(t)\rangle^{2}}.
\end{align}

## Sistemas bi-dimensionais
Muitas vezes é útil, ou uma boa aproximação, considerarmos que os sistemas físicos estudados estão restritos a se movimentar em uma ou em duas dimensões. Para um sistema bi-dimensional restrito a se mover na coordenada $z=0$, podemos escrever
\begin{equation}
\Psi(x,y,z,t)\equiv \Lambda(x,y,t)\delta(z),
\end{equation}
com a _função delta de Dirac_ satisfazendo 
\begin{equation}
\delta(z)=\begin{cases}0 \text{ se } z\ne 0 \\ \infty\text{ se } z =0\end{cases} \text{, }\delta(z)^{2}=\delta(z) \text{, } \int_{-\infty}^{+\infty}\delta(z)dz=1 \text{ e } \int f(w)\delta(w-w_{0})dx=f(w_{0}).
\end{equation}
Assim temos normalização
\begin{equation}
\iiint|\Psi(x,y,z,t)|^{2}dxdydz = \iint|\Lambda(x,y,t)|^{2}dxdy\int\delta(z)dz=\iint|\Lambda(x,y,t)|^{2}dxdy=1.
\end{equation}
Para os valores médios envolvendo a posição do quanton:
\begin{align}
\langle X(t)\rangle & = \iiint x|\Lambda(x,y,t)\delta(z)|^{2}dxdydz = \iint x|\Lambda(x,y,t)|^{2}dxdy\int\delta(z)dz \\
& = \iint x|\Lambda(x,y,t)|^{2}dxdy.
\end{align}
Da mesma forma $\langle Y(t)\rangle = \iint y|\Lambda(x,y,t)|^{2}dxdy$, e
\begin{equation}
\langle Z(t)\rangle = \iiint z|\Lambda(x,y,t)\delta(z)|^{2}dxdydz = \iint|\Lambda(x,y,t)|^{2}dxdy\int z\delta(z)dz=0.
\end{equation}

Para o momento linear teremos
\begin{align}
\langle P_{x}(t)\rangle & = \iiint \Lambda(x,y,t)^{*}\delta(z)^{*}\left(\frac{\hbar}{i}\partial_{x}\right)\Lambda(x,y,t)\delta(z)dxdydz = \iint \Lambda(x,y,t)^{*}\left(\frac{\hbar}{i}\partial_{x}\right)\Lambda(x,y,t)dxdy\int\delta(z)dz \\
& = \iint \Lambda(x,y,t)^{*}\left(\frac{\hbar}{i}\partial_{x}\right)\Lambda(x,y,t)dxdy.
\end{align}
Da mesma forma $\langle P_{y}(t)\rangle = \iint \Lambda(x,y,t)^{*}\left(\frac{\hbar}{i}\partial_{y}\right)\Lambda(x,y,t)dxdy$. Para
\begin{align}
\langle P_{z}(t)\rangle & = \iiint \Lambda(x,y,t)^{*}\delta(z)^{*}\left(\frac{\hbar}{i}\partial_{z}\right)\Lambda(x,y,t)\delta(z)dxdydz = \iint|\Lambda(x,y,t)|^{2}dxdy\int(\hbar/i)\delta(z)\partial_{z}\delta(z)dz \\
& = (\hbar/2i)\left(\delta(z)^{2}\right)_{-\infty}^{\infty} = 0.
\end{align}

OBS. A observação mais importante a ser deixada aqui é que se restringirmos uma coordenada do quanton a um certo valor, podemos ignorá-la nos cálculos que serão feitos usando somente $\Psi(x,y,t)$. Ficará como exercício verificar que podemos fazer isso também quando duas coordenadas são fixadas. Por exemplo, se fixamos $y$ e $z$ usaremos somente $\Psi(x,t)$ para fazer os cálculos.

# Partícula livre
A função de onda que consideramos acima,
\begin{equation}
\Psi_{d}(\vec{r},t)=Ae^{i(\vec{k}\cdot\vec{r}-\omega t)},
\end{equation}
pode ser vista como descrevendo um quanton livre,
\begin{equation}
U=0\therefore E=\frac{P^{2}}{2m}=\frac{\hbar^{2}k^{2}}{2m},
\end{equation}
que se move na diração $\vec{k}$ com velocidade $\omega/k$. Para verificar essa afirmação consideramos o movimento dos nós da parte real dessa função, $|A|\cos(\vec{k}\cdot\vec{r}-\omega t)$, que são obtidos de $\vec{k}\cdot\vec{r}-\omega t=n\pi/2$ com $n=1,2,3,\cdots$. Assim
\begin{equation}
\hat{k}\cdot\vec{r} = \frac{n\pi}{2k}+\frac{\omega}{k}t.
\end{equation}

Para essa função de onda, usando
\begin{equation}
\nabla_{\vec{r}}\Psi_{d} = (\hat{i}\partial_{x}+\hat{j}\partial_{y}+\hat{k}\partial_{z})Ae^{i(k_{x}x+k_{y}y+k_{z}z-\omega t)} = (\hat{i}ik_{x}+\hat{j}ik_{y}+\hat{k}ik_{z})\Psi_{d} = i\vec{k}\Psi_{d},
\end{equation}
obteremos o valor esperado do momento linear:
\begin{equation}
\langle\vec{P}(t)\rangle = \iiint \Psi_{d}^{*}\vec{P}\Psi_{d}d^{3}r = \iiint \Psi_{d}^{*}(\hbar/i)\vec{\nabla}_{\vec{r}}\Psi_{d}d^{3}r = \iiint \Psi_{d}^{*}(\hbar/i)(i\vec{k})\Psi_{d}d^{3}r=\hbar k\hat{k}=\sqrt{2mE}\hat{k}.
\end{equation}

Vale a pena mencionar que como o momento linear de um quanton com a função de onda acima é bem definido, $\hat{k}\cdot\vec{p}=\hbar k\therefore\sigma_{\hat{k}\cdot\vec{P}}=0$, sua posição é totalmente incerta $\sigma_{\hat{k}\cdot\vec{X}}=\infty$. Isso implicaria em uma extensão infinita para o quanton. No entanto, mesmo quando essa aproximação é razolável, como e.g. para um próton movimentando-se para se chocar com um núcleo atômico, entre a fonte do feixe de prótons e a posição do núcleo há uma separação finita. Além disso, como $|\Psi_{d}(\vec{r},t)|^{2}=|A|^{2}=\text{constante}$, a normalização da função de onda fornece $A=0$. Para fugir desse problema de normalização, usamos o chamado _método de normalização da caixa_. Para $x,y,z\in[-L,L]$ teremos $|A|=1/(2L)^{3/2}$. Assumindo $A\in\mathbb{R}$, vem que
\begin{equation}
\Psi_{d}(\vec{r},t)=\frac{1}{(2L)^{3/2}}e^{i(\vec{k}\cdot\vec{r}-\omega t)}.
\end{equation}

Como vimos em aulas anteriores, na realidade a solução para ter um quanton localizado em uma região finita do espaço é considerar a combinação linear de um número finito de ondas harmônicas, que fará com que $\sigma_{\hat{k}\cdot\vec{P}}$ e $\sigma_{\hat{k}\cdot\vec{X}}$ sejam finitos, respeitando sempre a relação de indeterminação de Heisenberg $\sigma_{\hat{k}\cdot\vec{P}}\sigma_{\hat{k}\cdot\vec{X}}\ge\hbar$.

# Condições sobre a função de onda
Para garantir que quantidades fisicamente observáveis não apresentem comportamento não físico, tais como e.g. divergências e descontinuidades, impomos algumas restrições sobre a função de onda. 

A função de onda $\Psi$ existe, é solução da equação de Schrödinger e deve satisfazer as seguintes condições:
1. $\Psi$ e deve ser contínua, unívoca e finita (em geral).
1. A Primeiras derivadas de $\Psi$ nas coordenadas espaciais devem ser contínuas, unívocas e finitas.
1. $\Psi$ deve ir pra $0$ rápido o suficiente quando as coordenadas espaciais tendem a $\infty$.

A condição 1 é imposta para que a densidade de probabilidade $|\Psi|^{2}$ não mude bruscamente, para que não tenha dois valores em um mesmo ponto do espaço e para que não seja infinita. <br>
OBS. Em alguns casos a função de onda pode ser infinita, desde que seja diferenciável e normalizável. A função delta é um exemplo de tal função.

Como a EqS envolve derivadas segundas nas coordenadas espaciais, pelos mesmos motivos da condição 1, impomos a condição 2 para que as derivadas segundas sejam bem comportadas. 

A condição 3 é aplicada para que integral de normalização convirja.

OBS. A EqSIT é uma eq. diferencial de segunda ordem. Ao resolvê-la aparecerão duas constantes de integração que serão determinadas usualmente especificando $\psi(\vec{r})$ e $\vec{\nabla}_{\vec{r}}\psi(\vec{r})$ em um, ou mais, pontos do espaço.
